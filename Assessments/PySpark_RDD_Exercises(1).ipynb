```python
print("Total Activities:", rdd.count())
print("Unique Activities:", rdd.distinct().count())
print("Most Common:", max(rdd.countByValue().items(), key=lambda x:x[1]))
print("Least Common:", min(rdd.countByValue().items(), key=lambda x:x[1]))
print("LOGIN %:", (rdd.filter(lambda x:x=="LOGIN").count()/rdd.count())*100)
print("VIDEO_PLAY %:", (rdd.filter(lambda x:x=="VIDEO_PLAY").count()/rdd.count())*100)
```
