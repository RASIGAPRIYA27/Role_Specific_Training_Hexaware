```python
from pyspark.sql.functions import *
```

```python
df = spark.read.format("csv") \
.option("header",True) \
.option("inferSchema",True) \
.load("file:/Workspace/Users/azuser7222_mml.local@karthikirisoutlook.onmicrosoft.com/customer_orders.csv")

df.show()
```

```python
df.printSchema()
```

```python
print("Total Records")

print(df.count())
```

```python
df = df.dropDuplicates()
display(df)

print("Duplicate Records Removed")

print(df.count())
```

```python
df = df.fillna({
    "status":"In Transit",
    "delay_days":0,
    "quantity":1,
    "price":0
})

df.display()
```

```python
df = df.withColumn("order_date",to_date(col("order_date")))

df = df.withColumn("expected_date",to_date(col("expected_date")))

df = df.withColumn("delivery_date",to_date(col("delivery_date")))

df.display()
```

```python
df = df.withColumn("status",when((col("status")=="Delayed") &(col("delay_days")>5),"Escalated").otherwise(col("status"))
)

df.select("customer_name","product_name","delay_days","status").show()
```

```python
df = df.withColumn("order_value",col("quantity") * col("price"))

df.select("customer_name","product_name","order_value").show()
```

```python
delayed = df.filter(col("delay_days")>0)

delayed.display()
```

```python
df.groupBy("city").count().show()
```

```python
delay_report = delayed.groupBy("customer_name").agg(count("order_id").alias("TotalOrders"),sum("delay_days").alias("TotalDelayDays"))

delay_report.show()
```

```python
region_report = delayed.groupBy("city").agg(count("order_id").alias("DelayedOrders"))

region_report.show()
```

```python
df.groupBy("status").count().show()
```

```python
df.createOrReplaceTempView("CustomerOrders")
```

```python
%sql
SELECT * FROM CustomerOrders
LIMIT 10;
```

```python
%sql
SELECT customer_name,SUM(delay_days) AS TotalDelay
FROM CustomerOrders
GROUP BY customer_name
ORDER BY TotalDelay DESC
LIMIT 5;
```

```python
%sql
SELECT status,
COUNT(*) TotalOrders
FROM CustomerOrders
GROUP BY status;
```

```python
df.write.mode("overwrite").format("delta").save("/FileStore/delta/CustomerOrders")
```

```python
delta_df = spark.read.format("delta").load("/FileStore/delta/CustomerOrders")

delta_df.display()
```

```python
df.write.mode("overwrite").option("header",True).csv("/FileStore/output/CustomerOrdersCSV")
```

```python
csv_df = spark.read.csv("/FileStore/output/CustomerOrdersCSV",header=True,inferSchema=True)

csv_df.display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("CustomerOrderTable")
```

```python
%sql
SELECT customer_name,product_name,delay_days,status
FROM CustomerOrderTable
WHERE delay_days>0
ORDER BY delay_days DESC;
```
