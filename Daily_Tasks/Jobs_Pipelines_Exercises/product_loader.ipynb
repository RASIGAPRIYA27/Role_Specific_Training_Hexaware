```python
product_data = [
("P101","Laptop",65000),
("P102","Mobile",25000),
("P103","TV",45000),
("P104","Shoes",4000),
("P105","Watch",8000)
]

columns = [
"product_id",
"product_name",
"price"
]
```

```python
products_df = spark.createDataFrame(product_data,columns)
```

```python
display(products_df)
```

```python
products_df.createOrReplaceTempView("products")
```

```python
spark.sql("""
SELECT AVG(price) AS average_price FROM products
""").show()
```
