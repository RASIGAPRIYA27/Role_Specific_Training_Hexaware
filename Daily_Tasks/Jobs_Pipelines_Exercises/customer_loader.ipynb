```python
customer_data = [
(1,"Rahul","Hyderabad"),
(2,"Priya","Bangalore"),
(3,"Amit","Mumbai"),
(4,"Sneha","Hyderabad"),
(5,"Farhan","Delhi")
]

columns=[
"customer_id",
"name",
"city"
]
```

```python
customers_df = spark.createDataFrame(customer_data,columns)
```

```python
customers_df.write.mode("overwrite").saveAsTable("customer_table")
print("Customer Table Created")
```
