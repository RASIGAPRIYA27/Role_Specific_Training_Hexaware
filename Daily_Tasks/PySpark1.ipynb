```python
from pyspark.sql.functions import when
new=df.withColumn(
    "salary_category",
    when(col("salary")>=85000,"High")
    .when(col("salary")>=75000,"Medium")
    .otherwise("Low")
)

new.show()
```

```python
from pyspark.sql.functions import when, col

df = df.withColumn(
    "salary_category",
    when(col("salary") >= 85000, "High")
    .when(col("salary") >= 75000, "Medium")
    .otherwise("Low")
)
df.groupBy("salary_category").count().show()
```

```python
df = df.withColumn("bonus", col("salary") * 0.10)
df = df.withColumn("annual_salary", col("salary") * 12)
report_df = df.select(
    "employee_id",
    "name",
    "department",
    "salary",
    "bonus",
    "annual_salary"
)

report_df.show()
```
