```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("RDD").getOrCreate()
```

```python
numbers=[10,20,30,40,50]

rdd=spark.sparkContext.parallelize(numbers)

print(rdd.collect())
```

```python
rdd.count()
```

```python
rdd.take(4)
```

```python
result=rdd.map(
    lambda x:x*2
)

print(result.collect())
```

```python
result=rdd.map(
    lambda x:x*x
)

print(result.collect())
```

```python
num=[1,2,3,4,5,6,7,8,9,22,27]

rdd=spark.sparkContext.parallelize(num)

result=rdd.filter(lambda x:x%2==0)
print(result.collect())
```

```python
num=[10,10,20,30,40,40,50,50,60]

rdd=spark.sparkContext.parallelize(num)

result=rdd.distinct()
print(result.collect())
```

```python
rdd1=spark.sparkContext.parallelize([1,2,3])
rdd2=spark.sparkContext.parallelize([4,5,6])

result=rdd1.union(rdd2)

print(result.collect())
```

```python
num=[50,60,20,30,10]

rdd=spark.sparkContext.parallelize(num)

result=rdd.sortBy(
    lambda x:x
)
print(result.collect())
```

```python
num=[10,20,30,40]

rdd=spark.sparkContext.parallelize(num)

result=rdd.reduce(
    lambda x,y:x+y
)

print(result)
```
