```python
df=spark.read.csv("flightBooking.csv",header=True,inferSchema=True)
```

```python
df.show()
```

```python
df.select("passenger_name","from_city","to_city").show()
```

```python
df.filter(df.status=="Confirmed").show()
```

```python
df.filter(df.status=="Cancelled").show()
```

```python
df.filter(df.status=="Pending").show()
```

```python
df.filter(df.ticket_price>20000).show()
```

```python
df.filter(df.to_city.isin("Dubai","Singapore","London")).show()
```

```python
df.count()
```

```python
df.groupBy("airline").count().show()
```

```python
df.groupBy("status").count().show()
```

```python
from pyspark.sql.functions import sum

df.filter(df.status=="Confirmed").agg(sum("ticket_price")).show()
```

```python
from pyspark.sql.functions import avg

df.groupBy("airline").agg(avg("ticket_price")).show()
```

```python
from pyspark.sql.functions import max

df.agg(max("ticket_price")).show()
```

```python
from pyspark.sql.functions import min

df.agg(min("ticket_price")).show()
```

```python
df.sort(df.ticket_price.desc()).show()
```

```python
from pyspark.sql.functions import col

df=df.withColumn(
    "tax",col('ticket_price')*0.05
)

df.show()
```

```python
df=df.withColumn(
    "final_price",col("ticket_price")+col("tax")
)

df.show()
```

```python
from pyspark.sql.functions import when

df=df.withColumn(
    "price_category",
    when(col("ticket_price") >=30000,"Premium")
    .when(col("ticket_price") >=10000,"Standard")
    .otherwise("Budget")
)

df.show()
```

```python
df.groupBy("price_category").count().show()
```

```python
confirmed_df = df.filter(df["status"] == "Confirmed")

confirmed_df.write.mode("overwrite").parquet("confirmed_flights.parquet")

confirmed_df.show()
```
