```python
from pyspark.sql.functions import count,when,col

df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

```python
remove_null=df.na.drop()

remove_null.show()
```

```python
df.na.drop(subset=["booking_amount"]).show()
```

```python
remove_null=df.na.drop(
    subset=[
        "customer_name","service_type","booking_amount"
    ]
)

remove_null.show()
```

```python
null_handled=df.na.fill(
    {
        "city":"Unknown",
        "provider":"Not Available",
        "service_type" :"Unknown",
        "booking_status":"Unknown",
        "payment_mode":"Not Provided",
        "booking_amount":0
    }
)

null_handled.show()
```

```python
df=df.withColumn(
    "data_quality_status",when(
        col("city").isNull() |
        col("provider").isNull() |
        col("booking_amount").isNull() |
        col("booking_status").isNull(),
        "Incomplete"
    ).otherwise("Complete")
)

df.show()
```

```python
df.groupBy("data_quality_status").count().show()
```

```python
df.filter(df.data_quality_status=="Complete").show()
```

```python
df.filter(df.data_quality_status=="Incomplete").show()
```

```python
df=df.withColumn(
    "tax",col("booking_amount")*0.05
)

df.show()
```

```python
df=df.withColumn(
    "final_amount",col("booking_amount")+col("tax")
)

df.show()
```

```python
from pyspark.sql.functions import sum,avg

df.filter(df.booking_status=="Confirmed").agg(sum("final_amount")).show()
```

```python
dp=df.dropna(subset=["booking_amount"]).agg(avg("booking_amount")).show()
```

```python
avg_fill = df.agg(avg("booking_amount")).collect()[0][0]

avg_drop =df.na.drop(
    subset=["booking_amount"]
).agg(
    avg("booking_amount")
).collect()[0][0]

print("Average after fill:", avg_fill)
print("Average after drop:", avg_drop)
```
