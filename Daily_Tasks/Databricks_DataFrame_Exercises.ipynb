```python
df=df.withColumn(
    "row_number",row_number().over(window_spec)
)
df.show()
```

```python
from pyspark.sql.window import Window

window_spec=Window.orderBy(col("total_amount").desc())

df=df.withColumn(
    "rank",rank().over(window_spec)
)
df.show()
```

```python
df=df.withColumn(
    "dense_rank",dense_rank().over(window_spec)
)
df.show()
```

```python
df.withColumn(
    "rank",
    rank().over(window_spec)
).filter(
    col("rank") <= 5
).show()
```

```python
df.withColumn(
    "rank",
    rank().over(window_spec)
).filter(
    col("rank") == 1
).show()
```

```python
window = Window.partitionBy("city").orderBy(col("total_amount").desc())

df.withColumn(
    "rank",
    rank().over(window)
).filter(
    col("rank") == 1
).show()
```

```python
window = Window.partitionBy("category").orderBy(col("total_amount").desc())

df.withColumn(
    "rank",
    rank().over(window)
).filter(
    col("rank") == 1
).show()
```

```python
window = Window.orderBy("order_id")

df.withColumn(
    "running_total",
    sum("total_amount").over(window)
).show()
```

```python
window = Window.orderBy("order_id")

df.withColumn(
    "previous_revenue",
    lag("total_amount",1).over(window)
).show()
```

```python
window = Window.orderBy("order_id")

df.withColumn(
    "next_revenue",
    lead("total_amount",1).over(window)
).show()
```

```python
window = Window.orderBy("order_id")

df.withColumn(
    "previous_revenue",
    lag("total_amount").over(window)
).withColumn(
    "comparison",
    when(col("total_amount") > col("previous_revenue"),"Higher")
    .when(col("total_amount") < col("previous_revenue"),"Lower")
    .otherwise("Same")
).show()
```

```python
window = Window.orderBy("order_id")

df.withColumn(
    "previous_revenue",
    lag("total_amount").over(window)
).withColumn(
    "revenue_difference",
    col("total_amount") - col("previous_revenue")
).show()
```

```python
window = Window.partitionBy("city").orderBy(col("total_amount").desc())

df.withColumn(
    "rank",
    rank().over(window)
).filter(
    col("rank") <= 2
).show()
```

```python
window = Window.partitionBy("category").orderBy(col("total_amount").desc())

df.withColumn(
    "rank",
    rank().over(window)
).filter(
    col("rank") <= 3
).show()
```

```python
window = Window.orderBy(col("total_amount").desc())

df.withColumn(
    "rank",
    rank().over(window)
).select(
    "order_id",
    "product",
    "city",
    "category",
    "total_amount",
    "rank"
).show()
```
