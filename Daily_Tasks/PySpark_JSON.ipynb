```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("Working With files").getOrCreate()
```

```python
from google.colab import files

uploaded=files.upload()
```

```python
df = spark.read.option("multiline","true").json("hotels.json")
```

```python
df.printSchema()
```

```python
df.show(truncate=False)
```

```python
df.select("hotel_name","city","rating").show()
```

```python
df.filter(df.category=="Luxury").show()
```

```python
df.filter(df.rating>4.5).show()
```

```python
df.filter(df.rooms_available>15).show()
```

```python
df.filter(df.price_per_night>=10000).show()
```

```python
df.filter((df.city=="Dubai")|(df.city=="London")).show()
```

```python
df.filter(df.contact.phone.isNull()).show()
```

```python
df.filter(df.contact.email.isNull()).show()
```

```python
df.select("hotel_name","contact.phone","contact.email").show()
```

```python
from pyspark.sql.functions import col,array_contains

df.filter(array_contains(col("amenities"), "wifi")).show()
```

```python
df.filter(array_contains(col("amenities"),"spa")).show(truncate=False)
```

```python
df.groupBy("city").count().show()
```

```python
df.groupBy("category").count().show()
```

```python
from pyspark.sql.functions import avg

df.groupBy("category").agg(avg("rating")).show()
```

```python
df.groupBy("city").agg(avg("price_per_night")).show()
```

```python
from pyspark.sql.functions import max,sum

df.agg(max("price_per_night")).show()
```

```python
df=df.withColumn(
    "total_potential_revenue",col("rooms_available") * col("price_per_night")
)

df.show(truncate=False)
```

```python
df.sort(df.rating.desc()).show(truncate=False)
```

```python
hotels_flat = df.select(
    "hotel_id","hotel_name","city","category","rating",
    "rooms_available","price_per_night",
    col("contact.phone").alias("phone"),
    col("contact.email").alias("email")
)

hotels_flat.write.mode("overwrite").parquet("hotels_flattened.parquet")
```
